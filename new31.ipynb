## באיזה מדינה לדעתכם רוב מסעדות המשלן בעולם לשנת 2019

```python
res_to_countries= pd.value_counts(Restaurant.region)
res_to_countries=res_to_countries.reset_index()
plt.figure(figsize = (15,8))
sns.barplot(data=res_to_countries, x='index',y='region')
plt.xticks(rotation=90)
plt.xlabel('Countries',fontsize=15)
plt.ylabel('Number of restaurants',fontsize=15)
plt.title("Number of restaurants for Countries",fontsize=30)
plt.show()
```

## לפי הגרף שמולנו, נוכל לראות שרוב מסעדות המישלן נמצאות באנגליה למרות ששיטת דירוג המסעדות על פי כוכבי מישלן התחילה בצרפת.
## בנוסף רוב מסעדות המישלן בעלות כוכב 1 נמצאות ביבשת אירופה

```python
res_to_city= pd.value_counts(Restaurant.city)
fifteen_max_city= res_to_city.nlargest(n=15)
fifteen_max=fifteen_max_city.reset_index()
plt.figure(figsize = (15,8))
sns.barplot(data=fifteen_max, x='index',y='city',ci=None)
plt.xticks(rotation=90)
plt.xlabel('city',fontsize=15)
plt.ylabel('Number of restaurants',fontsize=15)
plt.title("Number of restaurants for city",fontsize=25)
plt.show()
```

## אנו רואים שבערי מסחר עם אוכלוסיה גדולה ועשירה לרוב נמצא כמות מסעדות מישלן גבוהה

```python
res_to_cousines= pd.value_counts(Restaurant.cuisine)
ten_max_cousines= res_to_cousines.nlargest(n=10)
ten_max=ten_max_cousines.reset_index()
plt.figure(figsize = (15,8))
sns.barplot(data=ten_max, x='index',y='cuisine')
plt.xticks(rotation=90,fontsize=10)
plt.xlabel('cuisines',fontsize=15)
plt.ylabel('Number of restaurants',fontsize=15)
plt.title("Number of restaurants for cousines",fontsize=25)
plt.show()
```

```python
labels = ten_max['index']
sizes = ten_max['cuisine']
fig,ax = plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.2f%%',pctdistance=0.7, radius=3)
plt.show()
```

## אפשר לראות שהמטבח המודרני הוא הנפוץ ביותר בקרב מסעדות משלן בעולם

```python
datanonan= Restaurant.dropna(subset=['price'])
new= datanonan.iloc[:,[5,8]]
plt.figure(figsize = (15,6) )
plt.xticks(rotation=90,fontsize=13)
new = new.replace({'price': {'$':1, '$$':2, '$$$':3, '$$$$':4, '$$$$$':5}})
new =new.groupby("region")[["price"]].mean().round(0)
new =new.reset_index()
sns.barplot(data = new , y = 'price',x='region', hue='price')
plt.xlabel('regin',fontsize=15)
plt.ylabel('price',fontsize=15)
plt.title("avrege price for country",fontsize=25)
plt.show()
```

```python
price_for_cuisine= datanonan.iloc[:,[5,7,8]]
price_for_cuisine = price_for_cuisine.replace({'price': {'$':1, '$$':2, '$$$':3, '$$$$':4, '$$$$$':5}})
price_for_cuisine =price_for_cuisine.groupby("cuisine")[["price"]].mean().round(0)
price_for_cuisine =price_for_cuisine.reset_index()
plt.figure(figsize = (5,25) )
plt.xticks(rotation=90)
sns.scatterplot(data = price_for_cuisine, x = 'price', y = 'cuisine')
plt.title("avrege price for cuisine",fontsize=25)
plt.show()
```
